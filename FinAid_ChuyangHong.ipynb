```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
from subprocess import check_output
data = pd.read_excel('FinancialAid_1.xlsx',header=0)
row,col=data.shape
years=[]
value_pri=[]
value_pub=[]

for i in range(2,22):
        value_pri.append(data.iloc[i,4])
        value_pub.append(data.iloc[i,9])
        
for i in range(1997,2017):
        years.append(i)

plt.plot(years,value_pri,label="Undergraduate Students")
plt.plot(years,value_pub,label="Graduate students")


plt.xlabel("Years")
plt.ylabel("Financial Aid/Dollor")
plt.title("Finacial Aid per student")
plt.legend(loc=0)
plt.show()
```
