```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
from subprocess import check_output
#data = pd.read_excel('Employment_1.xls',header=0,usecols=range())
data = pd.read_excel('Tuition_1.xlsx',header=0)
row,col=data.shape
years=[]
value_pri=[]
value_pub=[]

for i in range(21,49):
        value_pri.append(data.iloc[i,7])
        value_pub.append(data.iloc[i,9])
        
for i in range(1990,2018):
        years.append(i)

plt.plot(years,value_pri,label="Private Nonprofit")
plt.plot(years,value_pub,label="Public")


plt.xlabel("Years")
plt.ylabel("Tuition and Fees and Room and Board/Dollar")
plt.title("Tuition and Fees and Room and Board,four-years")
plt.legend(loc=0)
plt.show()
```
