```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
from subprocess import check_output
#data = pd.read_excel('Employment_1.xls',header=0,usecols=range())
data = pd.read_excel('Salary_2.xls',header=0)
row,col=data.shape
years=[]
value_25_lh=[]
value_25_h=[]
value_25_c=[]
value_25_a=[]
value_25_b=[]
value_25_g=[]
for i in range(col):
    if (i-1)%3==0:
        s1=int(data.iloc[1,i])
        years.append(s1)
        value_25_lh.append(data.iloc[6,i])
        value_25_h.append(data.iloc[7,i])
        value_25_c.append(data.iloc[8,i])
        value_25_a.append(data.iloc[9,i])
        value_25_b.append(data.iloc[11,i])
        value_25_g.append(data.iloc[12,i])
figsize=10,5


plt.plot(years,value_25_lh,label="less than high school")
plt.plot(years,value_25_h,label="high school completion")
plt.plot(years,value_25_c,label="Some college")
plt.plot(years,value_25_a,label="Associate's degree")
plt.plot(years,value_25_b,label="Bachelor's degree")
plt.plot(years,value_25_g,label="Master's/PhD degree")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Anual earning for person 25-34 years old")
plt.legend(loc=0)
plt.show()
```
