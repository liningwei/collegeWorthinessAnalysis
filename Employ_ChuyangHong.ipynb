```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
from subprocess import check_output
#data = pd.read_excel('Employment_1.xls',header=0,usecols=range())
data = pd.read_excel('Employment_1.xls',header=0)
row,col=data.shape
years=[]
value_20_lh=[]
value_20_h=[]
value_20_c=[]
value_20_b=[]
value_25_lh=[]
value_25_h=[]
value_25_c=[]
value_25_b=[]
for i in range(10,col):
    if (i-1)%3==0:
        s1=int(data.iloc[1,i])
        years.append(s1)
        value_20_lh.append(data.iloc[9,i])
        value_20_h.append(data.iloc[10,i])
        value_20_c.append(data.iloc[11,i])
        value_20_b.append(data.iloc[12,i])
        value_25_lh.append(data.iloc[15,i])
        value_25_h.append(data.iloc[16,i])
        value_25_c.append(data.iloc[17,i])
        value_25_b.append(data.iloc[18,i])
figsize=10,5
plt.subplots(figsize=figsize)
ax1=plt.subplot(1,2,1)
plt.plot(years,value_20_lh,label="less than high school")
plt.plot(years,value_20_h,label="high school completion")
plt.plot(years,value_20_c,label="Some college")
plt.plot(years,value_20_b,label="Bachelor degree")

plt.xlabel("Years")
plt.ylabel("Employment rate/%")
plt.title("Employment rate for person 20-24 years old")
plt.legend(loc=0)

ax1=plt.subplot(1,2,2)
plt.plot(years,value_25_lh,label="less than high school")
plt.plot(years,value_25_h,label="high school completion")
plt.plot(years,value_25_c,label="Some college")
plt.plot(years,value_25_b,label="Bachelor degree")

plt.xlabel("Years")
plt.ylabel("Employment rate/%")
plt.title("Employment rate for person 25-64 years old")
plt.legend(loc=0)
plt.show()
```
